# Disease Prediction using ML


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("Training.csv")
```

```python
train_data.head(20)
```

```python
data_frame = train_data.dropna(axis=1)
```

```python
data_frame.head(20)
```

```python
test = pd.read_csv("Testing.csv")
```

```python
test.head(20)
```

```python
# Now, Lets split the features and labels of train data and test data 

features_train = data_frame.iloc[:, :-1].values
labels_train = data_frame.iloc[:, -1].values

features_test = test.iloc[:, :-1].values
labels_test = test.iloc[:, -1].values
```

```python
features_train = pd.DataFrame(features_train)
labels_train = pd.DataFrame(labels_train)

features_test = pd.DataFrame(features_test)
labels_test = pd.DataFrame(labels_test)
```

```python
features_train.head(20)
```

```python
labels_train.head(20)
```

```python
features_test.head(20)
```

```python
labels_test.head(20)
```

```python
# Lets Encode the labels in train and test data

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

labels_train = encoder.fit_transform(data_frame.iloc[:, -1])
labels_test = encoder.transform(test.iloc[:, -1])
```

```python
# Encoded train labels

labels_train = pd.DataFrame(labels_train)
labels_train.head(20)
```

```python
# Encoded test labels

labels_test = pd.DataFrame(labels_test)
labels_test.head(20)
```

```python
# Reshape test labels to a 1d array using ravel()

from numpy import ravel

labels_train = ravel(labels_train)
labels_test = ravel(labels_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

### Implementing Logistic Regression for Disease Prediction

```python
# Lets implement Logistic Regression

from sklearn.linear_model import LogisticRegression

LOGISTIC = LogisticRegression()
```

```python
# Lets prepare LR model using training features and labels

LOGISTIC.fit(features_train, labels_train)
```

```python
# Lets predict the test labels for test features

LR_test_labels_pred = LOGISTIC.predict(features_test)
```

```python
# Check the accuracy

LOGISTIC_accuracy = accuracy_score(labels_test, LR_test_labels_pred)

print("LOGISTIC_REGRESSION ACCURACY RESULT :- ", LOGISTIC_accuracy)
```

```python
# Precision

LOGISTIC_precision = precision_score(labels_test, LR_test_labels_pred, average = 'macro')

print("LOGISTIC_REGRESSION PRECISION RESULT :- ", LOGISTIC_precision)
```

```python
# Recall

LOGISTIC_recall = recall_score(labels_test, LR_test_labels_pred, average='macro')

print("LOGISTIC_REGRESSION ACCURACY RESULT :- ", LOGISTIC_recall)
```

```python
# F1 - Score

LOGISTIC_f1_score = f1_score(labels_test, LR_test_labels_pred, average='macro')

print("LOGISTIC_REGRESSION ACCURACY RESULT :- ", LOGISTIC_f1_score)
```

### Implementing Support Vector Machine

```python
# First, Lets Implement Support Vector Machine

from sklearn.svm import SVC

SVM = SVC(kernel = 'linear')
```

```python
# Lets prepare SVM model using training features and labels

SVM.fit(features_train, labels_train)
```

```python
# Lets predict the test labels for test features

SVM_TestLabels_prediction = SVM.predict(features_test)
```

```python
# Check the accuracy

SVM_accuracy_check = accuracy_score(labels_test, SVM_TestLabels_prediction)

print("SVM Accuracy :- : ", SVM_accuracy_check)
```

```python
# Precision

SVM_precision_check = precision_score(labels_test, SVM_TestLabels_prediction, average = 'macro')

print("SVM Precision :- ", SVM_precision_check)
```

```python
# Recall

SVM_recall_check = recall_score(labels_test, SVM_TestLabels_prediction, average='macro')

print("SVM Recall :- ", SVM_recall_check)
```

```python
# F1 - Score

SVM_f1_score_check = f1_score(labels_test, SVM_TestLabels_prediction, average='macro')

print("SVM F1-score :- ", SVM_f1_score_check)
```

### Implementing Decision Tree Algorithm

```python
# Lets implement Decision Tree

from sklearn.tree import DecisionTreeClassifier
Decision_T = DecisionTreeClassifier()
Decision_T.fit(features_train, labels_train)
```

```python
DT_TestlLabels_prediction = Decision_T.predict(features_test)
```

```python
# Accuracy

Decision_T_accuracy = accuracy_score(labels_test, DT_TestlLabels_prediction)

print("DT Accuracy :- ", Decision_T_accuracy)
```

```python
# Precision

Decision_T_precision = precision_score(labels_test, DT_TestlLabels_prediction, average = 'macro')

print("DT Precision :- ", Decision_T_precision)
```

```python
# Recall

Decision_T_recall = recall_score(labels_test, DT_TestlLabels_prediction, average='macro')

print("DT Recall:- ", Decision_T_recall)
```

```python
# F1 - Score

Decision_T_f1_score = f1_score(labels_test, DT_TestlLabels_prediction, average='macro')

print("DT F1_score :- ", Decision_T_f1_score)
```

### COMPARING SCORES OF SVM, DT AND LR

```python
# Accuracy

names = ['SVM', 'LR', 'DT']
scores = [SVM_accuracy_check, LOGISTIC_accuracy, Decision_T_accuracy]

plt.bar(names, scores)
plt.title("Comparing Accuracy of differnet Supervised Learning algorithms")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.show()
```

```python
# Precision

names = ['SVM', 'LR', 'DT']
scores = [SVM_precision_check, LOGISTIC_precision, Decision_T_precision]

plt.bar(names, scores)
plt.title("Comparing Accuracy of differnet Supervised Learning algorithms")
plt.xlabel("Model")
plt.ylabel("Precison")
plt.show()
```

```python
# RECALL

names = ['SVM', 'LR', 'DT']
scores = [SVM_recall_check, LOGISTIC_recall, Decision_T_recall]

plt.bar(names, scores)
plt.title("RECALL COMPARISON of VARIOUS MODELS")
plt.xlabel("Model")
plt.ylabel("Recall")
plt.show()
```

```python
# F1 - Score

names = ['SVM', 'LR', 'DT']
scores = [SVM_f1_score_check, LOGISTIC_f1_score, Decision_T_f1_score]

plt.bar(names, scores)
plt.title("F1 - Score COMPARISON of VARIOUS MODELS")
plt.xlabel("Model")
plt.ylabel("F1 - Score")
plt.show()
```
